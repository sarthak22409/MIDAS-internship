```python
def extract_features(file_name,**kwargs):    #extract features from audio file name such as contrast,tonnetz,mel,mfcc,chroma
     
        
    mfcc = kwargs.get("mfcc")     #representation of power spectrum of a sound bas ed on linear cosine transform
    chroma=kwargs.get("chroma")   #closely releates to twelve different pitch classes
    mel = kwargs.get("mel")
    contrast = kwargs.get("contrast") #Contrast Analysis can analyze two different selections in an audio track so as to determine the RMS difference (contrast) between foreground (speech) and background (music, audience noise or similar). The purpose is to determine if the speech will be intelligible to the hard of hearing.
    tonnetz = kwargs.get("tonnetz")
    with soundfile.SoundFile(file_name) as sound_file:
        X=sound_file.read(dtype = "float32")
        sample_rate = sound_file.samplerate   #to extract frequencies in audible range of human 
        if contrast or chroma:
            stft = np.abs(librosa.stft(X))
        result = np.array([])
        if mfcc:
            mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40).T, axis=0)
            result = np.hstack((result, mfccs))
                
        if chroma:
            chroma = np.mean(librosa.feature.mfcc(y=X,sr = sample_rate,n_mfcc=40).T,axis = 0)
            result = np.hstack((result,chroma))
        if mel:
            mel = np.mean(librosa.feature.melspectrogram(X, sr=sample_rate).T,axis=0)
            result = np.hstack((result, mel))
        if contrast:
            contrast = np.mean(librosa.feature.spectral_contrast(S=stft, sr=sample_rate).T,axis=0)
            result = np.hstack((result, contrast))
        if tonnetz:
            tonnetz = np.mean(librosa.feature.tonnetz(y=librosa.effects.harmonic(X), sr=sample_rate).T,axis=0)
            result = np.hstack((result, tonnetz))
        return result
```

```python
def load_data1(test_size=0.2):
    X, y = [], []
    for file in glob.glob("all_emotions_1/*.wav"):
        # get the base name of the audio file
        basename = os.path.basename(file)
        # get the emotion label
        basename = os.path.basename(file)
        a = basename.split("_")[4]
        emotion = a.split(".")[0]
    
        # we allow only AVAILABLE_EMOTIONS we set
        
        # extract speech features
        features = extract_features(file, mfcc=True, chroma=True, mel=True)
        # add to data
        X.append(features)
        y.append(emotion)
    # split the data to training and testing and return it
    return train_test_split(np.array(X), y, test_size=test_size, random_state=7)
```

```python
int2emotion = {
    "01": "neutral",
    "02": "calm",
    "03": "happy",
    "04": "sad",
    "05": "angry",
    "06": "fearful",
    "07": "disgust",
    "08": "surprised"
}

# we allow only these emotions ( feel free to tune this on your need )
AVAILABLE_EMOTIONS = {
    "disgust",
    "fear",
    "sad",
    "neutral",
    "happy"
}

def load_data(test_size=0.2):
    X, y = [], []
    for file in glob.glob("data/Actor_*/*.wav"):
        # get the base name of the audio file
        basename = os.path.basename(file)
        # get the emotion label
        emotion = int2emotion[basename.split("-")[2]]
        # we allow only AVAILABLE_EMOTIONS we set
        if emotion not in AVAILABLE_EMOTIONS:
            continue
        # extract speech features
        features = extract_features(file, mfcc=True, chroma=True, mel=True)
        # add to data
        X.append(features)
        y.append(emotion)
    # split the data to training and testing and return it
    return train_test_split(np.array(X), y, test_size=test_size, random_state=7)
        
```

```python
# load dataset, 75% training 25% testing
X_train, X_test, y_train, y_test = load_data(test_size=0.25)
```

```python
# best model, determined by a grid search
model_params = {
    'alpha': 0.01,
    'batch_size': 256,
    'epsilon': 1e-08, 
    'hidden_layer_sizes': (300,), 
    'learning_rate': 'adaptive', 
    'max_iter': 500, 
}
```

```python
# initialize Multi Layer Perceptron classifier
# with best parameters ( so far )
model = MLPClassifier(**model_params)
```

```python
# train the model
print("[*] Training the model...")
model.fit(X_train, y_train)
```

```python

y_pred = model.predict(X_test)

# calculate the accuracy
accuracy = accuracy_score(y_true=y_test, y_pred=y_pred)

print("Accuracy: {:.2f}%".format(accuracy*100))
```
